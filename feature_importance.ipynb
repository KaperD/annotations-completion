```python
with open('data/usage_loader.pickle', 'rb') as f:
    usage_loader = pickle.load(f)
```

```python
importance_sum = defaultdict(int)
iterations = 10
rnd = random.Random()
for i in range(iterations):
    print(i)
    X, y, feature_names = usage_loader.load(size=1000, state=random.randint(1, 10000), train_fraction=1)
    model = CatBoost()
    model.fit(X, y)
    for k in range(len(feature_names)):
        importance_sum[feature_names[k]] += model.model.feature_importances_[k]
for key in importance_sum:
    importance_sum[key] /= iterations
```
